```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
# x= scaler.fit_transform(x)
scaler.fit(df_test.drop('phishing',axis=1))
scaled_features = scaler.transform(df_test.drop('phishing',axis=1))
df_feat = pd.DataFrame(scaled_features,columns=df_test.columns[:-1])
df_feat.head()
```

```python
x=df_feat.copy()
y=df_test['phishing']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
```

```python
featureScores
```

```python
print(featureScores.nlargest(50,'Score')) #print 20 best features
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import RepeatedKFold
feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0))
feature_sel_model.fit(x,y)
feature_sel_model.get_support()
selected_feature = x.columns[(feature_sel_model.get_support())]
x = x[selected_feature]
```
